# **data processing**

```python
movies.iloc[0]['genres']
```

```python
def converter(l):
  if not isinstance(l, list):
    l=eval(l)
  result=[]
  for d in l :
    result.append(d['name'])
  return result
```

```python
movies
# converter(str(movies.iloc[0]['genres']))
```

```python
movies['genres']=movies['genres'].apply(converter)
```

```python
movies
```

```python
movies['keywords']=movies['keywords'].apply(converter)
```

```python
movies['cast'][0]
```

```python
def cast_converter(l):
  result=[]
  l=eval(l)
  if l is not None :
    for i in range(0,min(len(l),3)):
      d=l[i]
      result.append(d['name'])
  return result


```

```python
cast_converter(movies['cast'][0])
```

```python
movies['cast']=movies['cast'].apply(cast_converter)
```

```python
movies['overview']=movies['overview'].apply(lambda x : x.split())
```

```python
movies
```

```python
def fetch_director(l):
  result=[]
  l=eval(l)
  if l is not None :
    for d in l :
      if d['job']=='Director' :

        result.append(d['name'])
  return result
```

```python
movies.crew[0]
```

```python
fetch_director(movies.crew[0])
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies.head(1)
```

```python
movies['genres']=movies['genres'].apply(lambda x : [i.replace(' ','')for i in x ])
movies['keywords']=movies['keywords'].apply(lambda x : [i.replace(' ','')for i in x ])
movies['cast']=movies['cast'].apply(lambda x : [i.replace(' ','')for i in x ])
movies['crew']=movies['crew'].apply(lambda x : [i.replace(' ','')for i in x ])
```

```python
movies
```

```python
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : ' '.join(x))
```

```python
new_df.iloc[0,2]
```

# **stemming**

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return ' '.join(y)
```

```python
new_df['tags'][0]
```

```python
stem(new_df['tags'][0])
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
new_df
```

# **vectorization**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features=5000 , stop_words='english')
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out().tolist()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]

  movie_list=sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x : x[1])[1:6]
  for i in movie_list :
    print(new_df.iloc[i[0]]['title'])

```

```python
import pickle as pk
pk.dump(new_df,open("movies.pkl","wb"))
```

```python
pk.dump(similarity,open("similarity.pkl","wb"))
```
